```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
filepath = "Boxplot.csv"
data = pd.read_csv(filepath)
```

```python
print(data.head(100))
print(data['Teaching Method'][1])
```

```python
categories = []
table = {}
for index in data.index:
    method = data['Teaching Method'][index]
    score = data['Score'][index]
    if method not in categories:
        categories.append(method)
        table[method] = []      
    table[method].append(score)
```

```python
print(table)
```

```python
df = pd.DataFrame(table)

# Create a boxplot
plt.figure(figsize=(10, 6))
plt.boxplot(df.values, vert=True, labels=df.columns)

# Customize labels and title
plt.xlabel('Scores')
plt.title('Boxplot of Scores by Teaching Method')

# Show the plot
plt.show()
```
